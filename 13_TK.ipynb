# **プログラムの名称**
VS NPC RUSSIANROULETTE
## **概要**
このゲームは、二人のプレイヤーが参加するロシアンルーレットを題材にしたターン制のゲームです。プレイヤーは一名が実際のプレイヤーで、もう一名がNPC（コンピューター制御の対戦相手）となります。
ルールと進行
ライフの設定: プレイヤーとNPCはそれぞれ初期ライフが3つ設定されています。ライフが0になると、そのプレイヤーはゲームから脱落します。
行動の選択: プレイヤーのターンごとに、以下のアクションを選択することができます。
ロシアンルーレット: 引き金を引きます。弾倉に弾が入っていた場合、ライフが1つ減ります。弾が入っていなければ無事です。

手錠: 手錠を使用すると、自分の次のターンをスキップできます。手錠はゲーム内で1回しか使用できません。

虫眼鏡: 虫眼鏡を使用すると、次の弾倉に弾が入っているかどうかを確認できます。虫眼鏡もゲーム内で1回しか使用できません。

NPCの行動: NPCも同様に3つのアクションから選択します。NPCの行動はランダムに決定されます。

ゲームの終了: どちらかのライフが0になった時点でゲームは終了です。生き残ったプレイヤーが勝者となります。

## **操作方法**

アイテムや行動に対応する数字を入力するとそのアイテムを使用、また行動できます。ロシアンルーレットは１、手錠は２、虫眼鏡は３です。

## **制作者のコメント**

大きなアピールポイントは、アイテムを使った戦略性の高さです。虫眼鏡を使って次の弾倉の状況を確認することで、プレイヤーはリスクをより正確に評価でき、次の行動を慎重に決定することができます。また、手錠を使ってターンをスキップする戦術も、緊張感の高まるタイミングで重要な選択肢となります。この工夫により、単なる運試しではなく、戦略的なゲームプレイが楽しめるように設計されています。

## **作成時間**

作成時間：**約20時間**



```python
import random

class Player:
    def __init__(self, name):
        self.name = name
        self.life = 3  # ライフを3に設定
        self.handcuffs = 1  # 手錠を1つ持っている
        self.magnifying_glass = 1  # 虫眼鏡を1つ持っている
        self.skip_next_turn = False  # 次のターンをスキップするかどうかのフラグ

    def use_handcuffs(self):
        if self.handcuffs > 0:
            self.handcuffs -= 1
            return True
        return False

    def use_magnifying_glass(self):
        if self.magnifying_glass > 0:
            self.magnifying_glass -= 1
            return True
        return False

def russian_roulette():
    chamber = [0, 0, 0, 0, 0, 1]  # 1つの弾丸を含むリボルバー
    random.shuffle(chamber)
    return chamber

def player_turn(player, opponent, chamber):
    print(f"{player.name}のターンです。")
    action = input(f"選択肢 (1: ロシアンルーレット 2: 手錠を使う 3: 虫眼鏡を使う) : ")

    if action == '1':  # ロシアンルーレット
        print(f"{player.name}は引き金を引く…")
        if chamber[0] == 1:
            print(f"パーン！ {player.name}に弾丸が当たりました。")
            player.life -= 1
            print(f"{player.name}の残りライフ: {player.life}")
            if player.life <= 0:
                print(f"{player.name}は死亡しました。")
        else:
            print(f"カチッ！ {player.name}は無事です。")
        chamber.append(chamber.pop(0))  # 次の弾倉に移動
    elif action == '2':  # 手錠を使う
        if player.use_handcuffs():
            print(f"{player.name}は手錠を使い、自分の次のターンをスキップします。")
            player.skip_next_turn = True
            return True
        else:
            print("手錠を持っていません！ロシアンルーレットをします。")
            return player_turn(player, opponent, chamber)
    elif action == '3':  # 虫眼鏡を使う
        if player.use_magnifying_glass():
            print(f"{player.name}は虫眼鏡を使い、銃に弾が入っているか確認します。")
            if chamber[0] == 1:
                print("銃に弾が入っています！")
            else:
                print("銃に弾は入っていません。")
        else:
            print("虫眼鏡を持っていません！ロシアンルーレットをします。")
            return player_turn(player, opponent, chamber)
    else:
        print("無効な選択です。再度試してください。")
        return player_turn(player, opponent, chamber)
    return True  # 通常のターン進行

def main():
    player = Player("プレイヤー")
    npc = Player("NPC")

    chamber = russian_roulette()
    turn = 0  # 0: プレイヤーのターン, 1: NPCのターン

    while player.life > 0 and npc.life > 0:
        if turn == 0 and not player.skip_next_turn:
            turn_skip = not player_turn(player, npc, chamber)
            if not turn_skip:
                turn = 1
        elif turn == 0 and player.skip_next_turn:
            print("プレイヤーのターンはスキップされました。")
            player.skip_next_turn = False
            turn = 1
        elif turn == 1 and not npc.skip_next_turn:
            print("NPCのターンです。")
            npc_action = random.choice(['roulette', 'handcuffs', 'magnifying_glass'])  # NPCの動作をランダムに選ぶ

            if npc_action == 'handcuffs' and npc.use_handcuffs():
                print("NPCは手錠を使い、自分の次のターンをスキップします。")
                npc.skip_next_turn = True
            elif npc_action == 'magnifying_glass' and npc.use_magnifying_glass():
                print("NPCは虫眼鏡を使い、銃に弾が入っているか確認します。")
                if chamber[0] == 1:
                    print("銃に弾が入っています！")
                else:
                    print("銃に弾は入っていません。")
            else:
                print("NPCは引き金を引く…")
                if chamber[0] == 1:
                    print("パーン！ NPCに弾丸が当たりました。")
                    npc.life -= 1
                    print(f"NPCの残りライフ: {npc.life}")
                    if npc.life <= 0:
                        print("NPCは死亡しました。")
                else:
                    print("カチッ！ NPCは無事です。")
                chamber.append(chamber.pop(0))  # 次の弾倉に移動
            turn = 0
        elif turn == 1 and npc.skip_next_turn:
            print("NPCのターンはスキップされました。")
            npc.skip_next_turn = False
            turn = 0

    if player.life > 0:
        print("おめでとう！あなたの勝ちです！")
    else:
        print("残念でした、あなたは敗北しました。")

if __name__ == "__main__":
    main()
```
